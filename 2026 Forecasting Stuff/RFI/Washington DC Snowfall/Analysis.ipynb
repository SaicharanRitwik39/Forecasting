```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('dc_february_snowfall.csv', parse_dates=['season'])
```

```python
df['feb_snowfall_in'] = df['feb_snowfall_in'].replace('T', 0.01)
df['feb_snowfall_in'] = pd.to_numeric(df['feb_snowfall_in'])
```

#### Frequency Analysis. Bucket Wise Historical Occurrences.

```python
bucket_1 = df[df['feb_snowfall_in'] < 2].count()
bucket_2 = df[(df['feb_snowfall_in'] >= 2) & (df['feb_snowfall_in'] < 5)].count()
bucket_3 = df[(df['feb_snowfall_in'] >= 5) & (df['feb_snowfall_in'] < 8)].count()
bucket_4 = df[df['feb_snowfall_in'] >=8].count()
print(bucket_1)
print(bucket_2)
print(bucket_3)
print(bucket_4)
```

#### MCS.

```python
data = df["feb_snowfall_in"].dropna().values
```

```python
N = 100_000  
simulated_feb = np.random.choice(data, size=N, replace=True)
p_lt_2   = np.mean(simulated_feb < 2)
p_2_5    = np.mean((simulated_feb >= 2) & (simulated_feb < 5))
p_5_8    = np.mean((simulated_feb >= 5) & (simulated_feb < 8))
p_ge_8   = np.mean(simulated_feb >= 8)

p_lt_2, p_2_5, p_5_8, p_ge_8, sum([p_lt_2, p_2_5, p_5_8, p_ge_8])
```
