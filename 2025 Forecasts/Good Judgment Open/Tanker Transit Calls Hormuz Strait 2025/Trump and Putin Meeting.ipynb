```python
import pandas as pd
import datetime 
import numpy as np
```

```python
target_date = '2026-02-01'
target_date = pd.to_datetime(target_date)
current_date = datetime.datetime.now()
total_range_date = '2026-03-31'
total_range_date = pd.to_datetime(total_range_date)

polymarket_prob = 0.63
```

Uniform Model Assumption.

```python
polymarket_prob * (target_date-current_date)/(total_range_date-current_date)
```

Hazard Rate Model.

```python
lambda_est = -np.log(1-polymarket_prob)/(total_range_date-current_date).days
lambda_est
prob_by_target = 1 - np.exp(-lambda_est * (target_date-current_date).days)
prob_by_target
```
