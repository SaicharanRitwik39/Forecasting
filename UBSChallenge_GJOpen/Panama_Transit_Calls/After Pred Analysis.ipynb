```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('panama-canal-daily-trans_afterpred.csv')
data.head()
```

```python
data['DateTime'] = pd.to_datetime(data['DateTime'])
start_date = '2024-07-15'
end_date = '2024-08-14' 
df = data[(data['DateTime'] >= start_date) & (data['DateTime'] <= end_date)]
```

```python
df.head()
```

```python
df['Total Ships'] = df['Number of Cargo Ships'] + df['Number of Tanker Ships']
df.tail()
```

```python
df['Total Ships'].sum()
```
